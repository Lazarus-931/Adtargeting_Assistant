```python
import torch
import pandas as pd
from typing import Union, Dict, Any, List
import os

def load_pytorch_model(pt_path: str) -> Dict[str, Any]:
    """
    Load a PyTorch model (.pt file) and return its contents as a dictionary.
    
    Args:
        pt_path: Path to the .pt file
        
    Returns:
        Dictionary containing the loaded model state
    """
    try:
        # Load the PyTorch model
        model_data = torch.load("data.pt", map_location=torch.device('cpu'))
        return model_data
    except Exception as e:
        print(f"Error loading PyTorch model: {e}")
        return {}

def read_csv_file(csv_path: str, **kwargs) -> Union[pd.DataFrame, None]:
    """
    Read a CSV file and return its contents as a pandas DataFrame.
    
    Args:
        csv_path: Path to the CSV file
        **kwargs: Additional arguments to pass to pd.read_csv
        
    Returns:
        DataFrame containing the CSV data or None if an error occurs
    """
    try:
        # First verify the file exists
        if not os.path.exists(csv_path):
            print(f"File not found: {csv_path}")
            return None
            
        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(csv_path, **kwargs)
        
        # Verify successful reading with informative output
        print(f"Successfully read CSV file: {csv_path}")
        print(f"Shape: {df.shape} ({len(df)} rows, {len(df.columns)} columns)")
        print(f"Columns: {df.columns.tolist()}")
        print(f"First 3 rows preview:\n{df.head(3)}")
        
        return df
    except Exception as e:
        print(f"Error reading CSV file: {e}")
        return None

def extract_info_from_csv(csv_path: str) -> str:
    """
    Extract a text summary of a CSV file including column names and data samples.
    
    Args:
        csv_path: Path to the CSV file
        
    Returns:
        String containing information about the CSV file
    """
    try:
        df = pd.read_csv(csv_path)
        
        # Get basic information
        row_count = len(df)
        column_count = len(df.columns)
        column_names = list(df.columns)
        
        # Create a summary
        summary = f"CSV File Summary:\n"
        summary += f"Number of rows: {row_count}\n"
        summary += f"Number of columns: {column_count}\n"
        summary += f"Column names: {', '.join(column_names)}\n\n"
        
        # Add sample data (first 5 rows)
        if row_count > 0:
            summary += "Sample data (first 5 rows or fewer):\n"
            sample_rows = min(5, row_count)
            for i in range(sample_rows):
                summary += f"Row {i+1}: {dict(df.iloc[i])}\n"
        
        return summary
    except Exception as e:
        return f"Error processing CSV file: {e}"
    
    
    
def demographic_prompt(csv_path: str ,pt_path: str) -> str:
    """
    Generate a prompt for demographic analysis based on a CSV file and a PyTorch model.
    
    Args:
        csv_path: Path to the CSV file
        pt_path: Path to the PyTorch model (.pt file)
        
    Returns:
        A formatted prompt string
    """
    csv_info = extract_info_from_csv(csv_path)
    model_info = load_pytorch_model(pt_path)
    
    prompt = f"""
    You are an expert in demographic analysis. 
    The following CSV file contains data that you need to analyze:
    
    {csv_info}
    
    Additionally, you have access to a PyTorch model that can help with this analysis:
    
    {model_info}
    
    Please provide insights and recommendations based on the data provided.
    """
    

    
    return prompt

class InputState(TypeDict)
    question: str
```
